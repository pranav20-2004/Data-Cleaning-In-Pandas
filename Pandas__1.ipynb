Data Cleaning In
 pandas

```python
import pandas as pd
```

```python
df= pd.read_excel(r"/content/Customer Call List.xlsx")
df
```

First Remove Duplicate Values

```python
df.drop_duplicates()
```

To Update in Older DataSet

```python
df=df.drop_duplicates()
df
```

Drop The Column That Is Not Used

```python
df = df.drop(columns = "Not_Useful_Column")
df
```

```python
#df["Last_Name"]=df["Last_Name"].str.lstrip("...")
df
```

```python
#df["Last_Name"]=df["Last_Name"].str.strip("/")
#df["Last_Name"]=df["Last_Name"].str.rstrip("_")
df["Last_Name"]=df["Last_Name"].str.strip("123./_")
df
```

```python
#df["Phone_Number"]=df["Phone_Number"].str.replace('[^a-zA-Z0-9]','')
#df["Phone_Number"].apply(lambda x: x[0:3] + '-'+ x[3:6] + '-' + x[6:10])
#df["Phone_Number"].apply(lambda x: str(x))
df["Phone_Number"]=df["Phone_Number"].apply(lambda x: str(x)[0:3] + '-' + str(x)[3:6] + '-' + str(x)[6:10])
df
```

```python
df["Phone_Number"]=df["Phone_Number"].str.replace('nan--','')
df["Phone_Number"]=df["Phone_Number"].str.replace('Na--','')
df
```

```python
df[["Street Address","State","Zip Code"]]=df["Address"].str.split(',',expand=True)
df
```

```python
df=df.replace('Y','Yes')
df
```

```python
df["Paying Customer"]=df["Paying Customer"].str.replace('No','N')
df["Do_Not_Contact"]=df["Do_Not_Contact"].str.replace('No','N')
df

```

```python
df=df.replace('N/a','')
df["Do_Not_Contact"]=df["Do_Not_Contact"].str.replace('NaN','')
df
```

```python
df=df.fillna('')
df
```

```python
df["Paying Customer"]=df["Paying Customer"].str.replace('N','No')
df["Do_Not_Contact"]=df["Do_Not_Contact"].str.replace('N','No')
df
```

```python
for x in df.index:
  if df.loc[x,"Do_Not_Contact"] =="Yes":
    df.drop(x, inplace=True)
df
```

```python
for x in df.index:
  if df.loc[x,"Phone_Number"] =="":
    df.drop(x, inplace=True)
df
```

```python
df=df.reset_index(drop=True)
df
```
